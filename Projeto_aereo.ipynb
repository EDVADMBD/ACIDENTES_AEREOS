```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Dados fornecidos com campos expandidos
dados = {
    "Ano": [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025],
    "Acidentes": [172, 164, 146, 167, 151, 150, 141, 138, 155, 175, 22],
    "Mortes": [79, 104, 54, 81, 66, 55, 60, 49, 77, 152, 10],
    "Tipo de Aeronave": ["AviÃ£o", "HelicÃ³ptero", "AviÃ£o", "AviÃ£o", "HelicÃ³ptero", "AviÃ£o", "HelicÃ³ptero", "AviÃ£o", "AviÃ£o", "HelicÃ³ptero", "AviÃ£o"],
    "Local": ["AmÃ©rica", "Europa", "AmÃ©rica", "Ãsia", "AmÃ©rica", "Europa", "AmÃ©rica", "Ãsia", "AmÃ©rica", "Europa", "AmÃ©rica"],
    "Causa Principal": ["Erro Humano", "Falha MecÃ¢nica", "Erro Humano", "Clima", "Sabotagem", "Erro Humano", "Falha MecÃ¢nica", "Clima", "Erro Humano", "Falha MecÃ¢nica", "Clima"]
}

# Cria DataFrame
df = pd.DataFrame(dados)

# SimulaÃ§Ã£o de PrevenÃ§Ã£o
def simular_prevencao(anos):
    prevencoes = {
        "Erro Humano": 0,
        "Falha MecÃ¢nica": 0,
        "Clima": 0,
        "Sabotagem": 0
    }

    for ano in anos:
        causas = df[df['Ano'] == ano]['Causa Principal'].tolist()
        for causa in causas:
            if causa == 'Erro Humano':
                prevencoes['Erro Humano'] += random.randint(10, 20)  # Simula treinamento
            elif causa == 'Falha MecÃ¢nica':
                prevencoes['Falha MecÃ¢nica'] += random.randint(15, 25)  # Simula inspeÃ§Ãµes
            elif causa == 'Clima':
                prevencoes['Clima'] += random.randint(8, 15)  # Simula melhorias tecnolÃ³gicas
            elif causa == 'Sabotagem':
                prevencoes['Sabotagem'] += random.randint(5, 12)  # Simula reforÃ§o de seguranÃ§a

    return prevencoes

# Executar simulaÃ§Ã£o
anos_analise = [2023, 2024, 2025]
prevencoes = simular_prevencao(anos_analise)
print("\nSimulaÃ§Ã£o de PrevenÃ§Ã£o de Acidentes:\n", prevencoes)

# VisualizaÃ§Ã£o das medidas preventivas
plt.figure(figsize=(10, 5))
plt.bar(prevencoes.keys(), prevencoes.values(), color=['blue', 'green', 'orange', 'red'])
plt.title('Medidas Preventivas por Causa (SimulaÃ§Ã£o)')
plt.xlabel('Causa')
plt.ylabel('Investimentos Preventivos Simulados')
plt.show()

# RecomendaÃ§Ãµes Preventivas
print("\nRecomendaÃ§Ãµes Preventivas:")
print("- Erro Humano: Investir em treinamento e simulaÃ§Ãµes para pilotos e equipe de solo.")
print("- Falha MecÃ¢nica: Implementar manutenÃ§Ã£o preditiva e inspeÃ§Ãµes rigorosas.")
print("- Clima: Melhorar a previsÃ£o meteorolÃ³gica e tecnologias de monitoramento.")
print("- Sabotagem: ReforÃ§ar a seguranÃ§a aeroportuÃ¡ria e protocolos antiterrorismo.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Dados de acidentes aÃ©reos
dados_acidentes = {
    "Ano": list(range(2015, 2026)),
    "Acidentes": [172, 164, 146, 167, 151, 150, 141, 138, 155, 175, 22],
    "Mortes": [79, 104, 54, 81, 66, 55, 60, 49, 77, 152, 10]
}

df = pd.DataFrame(dados_acidentes)

# FunÃ§Ã£o para simular investimentos preventivos aleatÃ³rios
def simular_prevencao(df):
    prevencoes = []

    for _, row in df.iterrows():
        ano = row["Ano"]
        acidentes = row["Acidentes"]
        mortes = row["Mortes"]

        # SimulaÃ§Ã£o de investimentos aleatÃ³rios com base em proporÃ§Ãµes
        erro_humano = random.randint(50000, 150000) * (acidentes // 50)
        falha_mecanica = random.randint(100000, 300000) * (mortes // 20)
        clima = random.randint(30000, 100000) * (acidentes // 40)
        sabotagem = random.randint(20000, 80000) * (mortes // 30)

        prevencoes.append({
            "Ano": ano,
            "Erro Humano ($)": erro_humano,
            "Falha MecÃ¢nica ($)": falha_mecanica,
            "Clima ($)": clima,
            "Sabotagem ($)": sabotagem,
            "Total Investimentos ($)": erro_humano + falha_mecanica + clima + sabotagem
        })

    return pd.DataFrame(prevencoes)

# Gerar simulaÃ§Ã£o
df_prevencoes = simular_prevencao(df)

# Mostrar tabela com os dados simulados
print("Investimentos Preventivos Simulados:")
print(df_prevencoes)

# VisualizaÃ§Ã£o dos dados
plt.figure(figsize=(12, 6))
plt.plot(df_prevencoes["Ano"], df_prevencoes["Total Investimentos ($)"], marker='o', label='Total Investimentos')
plt.title("Investimentos Preventivos Simulados (2015-2025)")
plt.xlabel("Ano")
plt.ylabel("Investimentos ($)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Fatores de prevenÃ§Ã£o com seus respectivos valores simulados
investimentos_preventivos = {
    "Erro Humano": random.randint(500000, 1000000),
    "Falha MecÃ¢nica": random.randint(750000, 1500000),
    "Clima": random.randint(300000, 700000),
    "Sabotagem": random.randint(100000, 300000)
}

# Exibindo os investimentos
print("=== Investimentos Preventivos Simulados ===")
for fator, valor in investimentos_preventivos.items():
    print(f"{fator}: R${valor:,.2f}")

# SimulaÃ§Ã£o de reduÃ§Ã£o de riscos
riscos_iniciais = {"Erro Humano": 40, "Falha MecÃ¢nica": 35, "Clima": 15, "Sabotagem": 10}
reducoes_estimadas = {fator: round(valor / 1000000 * 5, 2) for fator, valor in investimentos_preventivos.items()}

print("\n=== ReduÃ§Ã£o Estimada de Riscos ===")
for fator, reducao in reducoes_estimadas.items():
    print(f"{fator}: {riscos_iniciais[fator] - reducao:.2f}% de risco restante")

# GrÃ¡fico de Investimentos
plt.figure(figsize=(10, 5))
plt.bar(investimentos_preventivos.keys(), investimentos_preventivos.values(), color='skyblue')
plt.title('Investimentos Preventivos por Fator')
plt.xlabel('Fatores de PrevenÃ§Ã£o')
plt.ylabel('Investimento (R$)')
plt.show()

# GrÃ¡fico de ReduÃ§Ã£o de Riscos
plt.figure(figsize=(10, 5))
plt.bar(riscos_iniciais.keys(), riscos_iniciais.values(), color='lightcoral', label='Risco Inicial')
plt.bar(reducoes_estimadas.keys(), [riscos_iniciais[fator] - reducao for fator, reducao in reducoes_estimadas.items()],
        color='lightgreen', label='Risco Reduzido')
plt.title('ReduÃ§Ã£o de Riscos Estimados por Fator')
plt.xlabel('Fatores de PrevenÃ§Ã£o')
plt.ylabel('Porcentagem de Risco')
plt.legend()
plt.show()
```
